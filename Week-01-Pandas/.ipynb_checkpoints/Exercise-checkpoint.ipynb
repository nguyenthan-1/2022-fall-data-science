# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
df.neighbourhood_group.value_counts()

```

```python
df.head()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
gb = df.groupby('neighbourhood_group')
gb['price'].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
gb['price'].agg(['min','max'])
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
groupby_cols = ['neighbourhood_group', 'room_type']
gb = df.groupby(groupby_cols)
gb['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
gb = df.groupby(groupby_cols)
gb['price'].agg(['min','max','count','max','mean','median','std'])
```

# Use the grammys.csv data for the next section of questions.

1. Who was won Album of the Year in 2016?


2. Who won Best Rap Album in 2009?


3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?

```python
grammys = pd.read_csv('data/grammys.csv')
grammys.head()
```

```python
#  1. Who was won Album of the Year in 2016?
condition1 = grammys['year'] == 2016
condition2 = grammys['category'] == 'Album of the Year'
condition3 = grammys['winner'] == True

grammys_2016 = grammys[condition1 & condition2 & condition3]
grammys_2016
```

```python
# 2. Who won Best Rap Album in 2009?
condition1 = grammys['year'] == 2009
condition2 = grammys['category'] == 'Best Rap Album'
condition3 = grammys['winner'] == True

grammys_2009 = grammys[condition1 & condition2 & condition3]
grammys_2009
```

```python
# 3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?
grammys[grammys['nominee'] == "Kendrick Lamar"]
```
